**Outlier Detection and Removal**

```python
# Detect and handle outliers (using IQR method)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

df = df[~((df < (Q1 - 1.5 * IQR)) | (df > (Q3 + 1.5 * IQR))).any(axis=1)]
```

**Feature Engineering**

```python
df['TAX_PER_ROOM'] = df['TAX'] / df['RM']
```
