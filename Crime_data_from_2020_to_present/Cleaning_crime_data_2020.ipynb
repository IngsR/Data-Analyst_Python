```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

Hal pertama yang perlu kita lakukan adalah memuat library dan dataset yang akan kita gunakan. Untuk demonstrasi ini, kita akan menggunakan Crime Data dari tahun 2020, yang berisi catatan kejadian kriminal yang telah dilaporkan di Amerika Serikat (US).

```python
df = pd.read_csv("../Data/Crime_Data_from_2020_to_Present.csv")

np.random.seed(0) 
df.info()
df.head()
```

```python
df.sample(5)
```

```python
missing_values = df.isna().sum()
missing_values = missing_values[missing_values > 0]
missing_values = missing_values.sort_values(ascending=False)

print(missing_values)
```

```python
# how many total missing values do we have?
missing_values_count = df.isnull().sum()
# Ganti np.product() dengan np.prod()
total_cells = np.prod(df.shape)
# Hitung total missing values
total_missing = missing_values_count.sum()
percent_missing = (total_missing / total_cells) * 100

print(f"Total missing values: {total_missing}")
print(f"Persentase missing values: {percent_missing:.2f}%")
```

```python
# remove all the rows that contain a missing value
df.dropna()
```

```python
columns_with_na_dropped = df.dropna(axis=1)
columns_with_na_dropped.head
```

```python
# just how much data did we lose?
print("Columns in original dataset: %d \n" % df.shape[1])
print("Columns with na's dropped: %d" % columns_with_na_dropped.shape[1])
```

```python
print(df.columns)
```

```python
df.dtypes
```

```python
df['Weapon Desc'] = df['Weapon Desc'].fillna('Not Listed')
df['Vict Sex'] = df['Vict Sex'].fillna('Unknown')
df['Vict Descent'] = df['Vict Descent'].fillna('Unknown')

df['Premis Desc'] = df['Premis Desc'].fillna(df['Premis Desc'].mode()[0])
df['Weapon Used Cd'] = df['Weapon Used Cd'].fillna(0)
df['Crm Cd 1'] = df['Crm Cd 1'].fillna(0)
df['Crm Cd 2'] = df['Crm Cd 2'].fillna(0)
df['Premis Cd'] = df['Premis Cd'].fillna(df['Premis Cd'].median())

# Tambahan untuk mengisi kolom lain yang masih memiliki missing values
df['Status'] = df['Status'].fillna(df['Status'].mode()[0])
df['Cross Street'] = df['Cross Street'].fillna('Unknown')
df['Mocodes'] = df['Mocodes'].fillna('Not Listed')

# Jika `Crm Cd 3` dan `Crm Cd 4` tidak terlalu penting, bisa dihapus
df.drop(columns=['Crm Cd 3', 'Crm Cd 4'], inplace=True)

# Mengecek kembali missing values setelah imputasi
missing_values = df.isna().sum()
missing_values = missing_values[missing_values > 0]
missing_values = missing_values.sort_values(ascending=False)

# Print hasil missing values setelah proses cleaning
print("Missing Values After Imputation:")
print(missing_values)
```

```python
print(df.isna().sum())
print("Totoal Missing values: %d" % df.isna().sum().sum())  
```

```python
df['DATE OCC'].sample(10)
```

```python
# Pastikan kolom 'DATE OCC' dalam format datetime
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'], errors='coerce', infer_datetime_format=True)
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'], errors='coerce', infer_datetime_format=True)

# Konversi kolom yang seharusnya integer
int_cols = ['Crm Cd 1', 'Crm Cd 2', 'Weapon Used Cd', 'Premis Cd']
df[int_cols] = df[int_cols].astype('Int64')

# Pastikan kolom kategori dalam format category untuk efisiensi
category_cols = ['AREA NAME', 'Crm Cd Desc', 'Weapon Desc', 'Status Desc', 
                 'Premis Desc', 'Vict Sex', 'Vict Descent', 'Status']
df[category_cols] = df[category_cols].astype('category')

# Cek ulang tipe data setelah konversi
print(df.dtypes)
```
